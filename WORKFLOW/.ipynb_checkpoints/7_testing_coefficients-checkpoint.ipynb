## Pipeline

Now, let's recreate the pipeline we had in the last notebook and then train the model.

```python
# the coefficients for all the features 
coefs = lr_all.coef_
coefs = coefs[0]
coefs
```

The coefficients are going to be log-odds, we need to convert them to odds in order to interpret them easier. 

What the coefficients are saying is this: Keeping all else  constant, if we increase a feature, 𝑥𝑖, by one unit, the log-odds of that feature increases by  𝛽𝑖 (coefficient).

This means that for every unit increase in  𝑥𝑖, the odds is multiplied  𝑒𝛽𝑖  times.

```python
coef_df = pd.DataFrame(data = coefs)
coef_df.columns = ["coef log-odds"]
coef_df["coef names"] = X_train_all.columns
coef_df["odds"] = coef_df['coef log-odds'].apply(lambda x: math.exp(x))
coef_df
```

```python
coef_df.columns
```

Note here that we set the y(odds) to be at most 3, because there is a feature (the object being from the library) that has an odds of nearly 400. But as earlier noted, this is a finickiness in data collection (ie. an object from the library is only included in this database if it is highlighted). 

Thus, we keep this feature in check and actually look at the rest of the features in detail. 

```python
plt.figure(dpi=150)
g = sns.barplot(coef_df['coef names'], coef_df['odds'])
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.ylim([0, 3])
g.set(xlabel= "Feature", ylabel = "Odds of a feature")
g.set_title('Odds of features being highlighted')
g.figure.savefig("Odds of features being highlighted", bbox_inches='tight'); 
```

```python
important_features = coef_df[coef_df.odds >1]
important_features
```

```python
plt.figure(dpi=150)
g = sns.barplot(important_features['coef names'], important_features['odds'])
g.set_xticklabels(g.get_xticklabels(), rotation=90)
plt.ylim([0, 3])
g.set(xlabel= "Feature", ylabel = "Odds of a feature")
g.set_title('Odds of features being highlighted')
g.figure.savefig("Odds of select features being highlighted", bbox_inches='tight'); 
```
